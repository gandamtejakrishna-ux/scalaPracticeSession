```scala
object Main {
  def main(args: Array[String]) = {
    val s = "Hello"
    val r = (1 to 5).map(_ * 2)
    val nums = Array(1, 2, 3)
    val set = Set(1, 2, 2, 3) // removes duplicates
    val map = Map("a" -> 1, "b" -> 2)
    // println(map("a" + "b"))
    println(s"$r")
    println(s"$s World")
    val ad = add(3, 9)
    val ad1 = add(10, 20)
    println(s"$ad $ad1")
    println(age(2))
    age1(List(10, 20, 30))
    println(age2())
    println(pal(12))
    println(swit(255))
    println(optionCase(Some("World")))
  }
  def add(a: Int, b: Int): Int = {
    val sub = a + b
    return sub
  }
  def age(c: Int): String = {
    val stat = if (c > 18) "Adult" else "Not Adult"
    // println(stat)
    return stat
  }
  def age1(d: List[Int]): Unit = {
    for (x <- d) {
      println(x)
    }
  }
  def age2(): List[Int] = {
    val num = for (x <- 1 to 5) yield x * 2
    num.toList
  }
  def palindrome(num: Int): String = {
    var n = num
    var obje = n
    var s = 0
    while (n > 0) {
      var d = n % 10
      s = s * 10 + d
      n = n / 10
    }
    if (s == obje) "Palindrome" else "Not Palindrome"
  }
  def swit(x: Int): String = {
    x match {
      case x if 10 > x => "Nice"
      case x if 20 > x => "good"
      case x if 30 > x => "okay"
      case _           => "fine"
    }
  }
  def optionCase(v: Option[String]): String = {
    v match {
      case Some(v) => s"Hello $v"
      case None    => "Hello no one"
    }
  }
}
```

```scala
// Variables

val name: String = "Alice"  // immutable
var counter: Int = 0            // mutable
counter = counter + 1
println(s"name=$name, counter=$counter")
```

```scala
// Datatypes
// Int, Long, Double, Float, Boolean, Char, String, Option[T], collections.

val i: Int = 42
val pi: Double = 3.14
val isOk: Boolean = true
val hello: Option[String]=Some("Hi")
println(hello.getOrElse(0))
val Hello="Hello There"
val maybe: Option[String] = Some(Hello)
val mayNotBe: Option[String] = None
println(s"i=$i, pi=$pi, isOk=$isOk, maybe=$maybe, mayNotBe=$mayNotBe")
```

```scala
// Decision-making statements

val x = 10
val result = if (x > 0) "positive" else "non-positive"
println(s"x=$x , result=$result")
```

```scala
// Control Structures

// while loop
var i = 0
while (i < 3) { 
    println(s"while: $i")
    i += 1 
}

// for-loop
for (j <- 1 to 3) 
println(s"for: $j")

for(x <- 1 to 10 if (x%2==0)){
    println(s"even number: $x")
}
```

```scala
// Arrays

val arr: Array[Int] = Array(1, 2, 3)
arr(0) = 10  // mutates first element
var arr1: Array[Int] = Array(4,5,6)
arr1(1)=20
println(s"Array after mutation: ${arr.mkString(",")}")
println(arr1.mkString("Array: [", ", ", "]"))
```

```scala
// Collections

val lst: List[Int] = List(1,2,3)
val vec: Vector[Int] = Vector(1,2,3)
val mp: Map[String, Int] = Map("a" -> 1)
println(s"List: ${lst}, Vector: ${vec}, Map: ${mp}")
```

```scala
// Functions — Basic concepts

def add(a: Int, b: Int): Int = a + b
val s = add(2,3)  // 5
println(s"add(2,3)=$s")
```

```scala
// Functions — Overloading

def greet(name: String): String = s"Hello, $name"
def greet(n: Int): String = s"Hello, user#$n"
println(greet("Alice"))
println(greet(42))
```

```scala
// Functions — Default parameters

def difference(x: Int, n: Int = 2): Int = {
    x-n
}
println(difference(3))    
println(difference(3, 3)) 
```

```scala
// Functions — Unit

def printMsg(msg: String): Unit = println(msg)
def printMsg2(msg: String) = println(msg)
printMsg("Hello from printMsg")
printMsg2("Hello from printMsg2")
```
